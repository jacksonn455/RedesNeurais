```
import numpy as np
import tensorflow as tf
import keras

from keras.preprocessing.image import load_img, img_to_array

#from keras.applications.inception_resnet_v2 import InceptionResNetV2, decode_predictions, preprocess_input
from keras.applications.vgg19 import VGG19, decode_predictions, preprocess_input

from IPython.core.display import display
```

```
imagem = load_img(ARQUIVO_3, target_size=(224, 224))
display(imagem)
```

```
imagem_array = img_to_array(imagem)
imagem_array.shape
```

```
modelo = VGG19(weights='imagenet')
```

```
modelo.graph = tf.compat.v1.get_default_graph()
```

```
array_expandido = np.expand_dims(imagem_array, axis=0)
array_expandido.shape
```

```
def formatar_modelo(nome_imagem):
  imagem = load_img(nome_imagem, target_size=(224, 224))
  imagem_array = img_to_array(imagem)
  array_expandido = np.expand_dims(imagem_array, axis=0)
  preprocessado = preprocess_input(array_expandido)

  return preprocessado
```

```
nome_imagem = ARQUIVO_14
imagem_2 = formatar_modelo(nome_imagem)
predicao = modelo.predict(imagem_2)
display(load_img(nome_imagem))
decode_predictions(predicao)
```
